```python

hashtags_numeric = []  # List to store the numeric representations of hashtags
hashtag_mapping = {}  # Dictionary to store unique hashtags and their corresponding numeric values

```

```python
for hashtag in hashtags_column:
    if hashtag not in hashtag_mapping:
        # Assign a new numeric value for each unique hashtag
        numeric_value = len(hashtag_mapping) + 1
        hashtag_mapping[hashtag] = numeric_value
    
    # Append the numeric value to the list
    hashtags_numeric.append(hashtag_mapping[hashtag])

hashtags_array = np.array(hashtags_numeric)  # Convert the list to a NumPy array

print(hashtags_array)
```
